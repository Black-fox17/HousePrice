```python
import pandas as pd
```

```python
data = pd.read_csv("House_price.csv")
```

```python
data[data["SalePrice"] > 500000]
```

```python
data.info()
```

```python
data["Location"].unique()
```

```python
data["Location"] = data["Location"].map({"Urban":0,"SubUrban":1,"Rural":2})
```

```python
data.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
input_data = data.drop(["SalePrice"],axis = 1)
#input_data["HouseAge"] = 2024 - input_data["HouseAge"]
output = data["SalePrice"]
```

```python
input_data
```

```python
x_train,x_val,y_train,y_val = train_test_split(input_data,output,train_size = 0.9)
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_val)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
score = mean_absolute_error(y_val,pred)
print(f"Loss Value is {score}")
```

```python
import pickle
with open("prediction.pkl","wb") as file:
    pickle.dump(model,file)
```
